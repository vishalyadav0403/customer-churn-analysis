```python
import pandas as pd
from sqlalchemy import create_engine

# =========================
# CONFIGURATION
# =========================

DB_TYPE = "postgresql"   # change to "mysql" if needed

USERNAME = "postgres"
PASSWORD = "vishalyadav33."
HOST = "localhost"
PORT = "5432"
DATABASE = "customer_churn_db"

# =========================
# LOAD DATA
# =========================

df = pd.read_csv("cleaned_telco_churn.csv")

# =========================
# DATABASE CONNECTION
# =========================

if DB_TYPE == "mysql":
    connection_string = (
        f"mysql+pymysql://{USERNAME}:{PASSWORD}@{HOST}:3306/{DATABASE}"
    )

elif DB_TYPE == "postgresql":
    connection_string = (
        f"postgresql+psycopg2://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}"
    )

else:
    raise ValueError("Unsupported database type!")

# Create engine
engine = create_engine(connection_string)

# =========================
# UPLOAD DATA
# =========================

df.to_sql(
    name="customer_churn",
    con=engine,
    if_exists="replace",
    index=False
)

print(f"Data uploaded successfully to {DB_TYPE}!")
```
